```python
from bokeh.plotting import figure
from bokeh.io import output_notebook
from bokeh.io import show

from bokeh.sampledata.iris import flowers as df
from bokeh.models import ColumnDataSource
```

```python
output_notebook()
```

```python
#Plot properties
```

```python
source = ColumnDataSource(df)

plot = figure(plot_width=400, plot_height=200)
plot.outline_line_width = 5
plot.outline_line_alpha = 0.5
plot.outline_line_color = "red"

plot.line(x='petal_length', y='petal_width', source=source)

show(plot)
```

```python
#Glyphs properties
```

```python
source = ColumnDataSource(df)

plot = figure(height=400, width=400)

circle_glyph = plot.circle('petal_length', 'petal_width', source=source)
circle_glyph.glyph.size = 10
circle_glyph.glyph.fill_alpha = 0.3
circle_glyph.glyph.line_color = "red"
circle_glyph.glyph.line_width = 2

show(plot)
```

```python
#Axes properties
```

```python
from math import pi

source = ColumnDataSource(df)

plot = figure(height=400, width=400)

plot.diamond('petal_length', 'petal_width', source=source)

plot.xaxis.axis_label = "petal length"
plot.xaxis.major_label_orientation = pi /4

plot.yaxis.axis_label = "petal width"
plot.yaxis.axis_label_text_color = "green"
plot.yaxis.major_label_orientation = "vertical"

show(plot)
```

```python
#Grids properties
```

```python
source = ColumnDataSource(df)

plot1 = figure(height=300, width=300)

plot1.x([1,2,3,4,5], [2,5,8,2,7], size=10, line_width=2)

plot1.xgrid.grid_line_alpha = 0.0

plot1.ygrid.grid_line_alpha = 0.5
plot1.ygrid.grid_line_dash = [6, 4]

show(plot1)
```
