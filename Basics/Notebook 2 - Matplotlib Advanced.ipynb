# Dual Y Axis 

```python
fig = plt.figure()

plot1 = fig.add_subplot(1, 2, 1)
plot1.plot([3, 4, 5], [10, 20, 30])
plot1.set_xlabel("Weeks")
plot1.set_ylabel("Sales")

plot2 = plot1.twinx()
plot2.plot([3, 4, 5], [50, 40, 30])
plt.show()
```

```python
fig = plt.figure()

plot1 = fig.add_subplot(1, 2, 1)
plot1.plot([3, 4, 5], [1, 2, 3])
plot1.set_xlabel("Weeks")
plot1.set_ylabel("Sales")

plot2 = plot1.twinx()
plot2.plot([3, 4, 5], [155, 122, 119])

plt.show()
```

```python
fig = plt.figure()

plot1 = fig.add_subplot(1, 2, 1)
plot1.plot([3, 4, 5], [1, 2, 3])
plot1.set_xlabel("Weeks")
plot1.set_ylabel("Sales")

plot2 = fig.add_subplot(1, 2, 2)
plot2.plot([3, 4, 5], [155, 122, 119])

plt.show()
```

# Logarithmic Axes

```python
fig = plt.figure()

plot1 = fig.add_subplot(2, 1, 1)
plot1.plot([3, 4, 5], [10, 20, 30])
plot1.set_xlabel("Weeks")
plot1.set_ylabel("Sales")

plot2 = fig.add_subplot(2, 1, 2)
plot2.plot([3, 4, 5], [10, 20, 30])
plot2.set_ylabel("log(sales)")
plot2.set_xlabel("Weeks")
plot2.set_yscale('log')
plt.show()
```

# Plotting Dates

```python
date1 = dt.datetime(2019, 4, 1, 12, 00, 00)
date2 = dt.datetime(2019, 4, 2, 12, 00, 00)
date3 = dt.datetime(2019, 4, 3, 12, 00, 00)

dates = [date1, date2, date3]
values = np.random.rand(len(dates))

plt.plot_date(mpl.dates.date2num(dates), values)

plt.show()
```

```python
date1 = dt.datetime(2019, 4, 1, 12, 00, 00)
date2 = dt.datetime(2019, 4, 2, 12, 00, 00)
date3 = dt.datetime(2019, 4, 3, 12, 00, 00)

dates = [date1, date2, date3]
values = np.random.rand(len(dates))

plot1 = plt.subplot(1, 1, 1)
plot1.plot_date(mpl.dates.date2num(dates), values)

dateFmt = mpl.dates.DateFormatter('%d-%m')
plot1.xaxis.set_major_formatter(dateFmt)

plt.show()
```
