# Notebook 1 - MatplotLib Basics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #using for builtin datasets
```

```python
plt.style.use('seaborn-white')
print(plt.style)
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

# Multiline Plot

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 5, 3, 2, 1, 2]
z = [4, 9, 3, 2, 8, 1]

plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

# Grids, Axes, Titles and Labels

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 5, 3, 2, 1, 2]

plt.plot(x,y)
plt.grid(False)
plt.axis(xmin = 0, xmax = 9, ymin = 0, ymax = 7)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 5, 3, 2, 1, 2]

plt.plot(x,y)
plt.xlabel("Age")
plt.ylabel("Sales amount")
plt.title("Sales amount by Age")
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 5, 3, 2, 1, 2]

plt.title("Legend Demo")

plt.plot(x,y, label="regular")
plt.plot(y,x, label="reverse")

plt.legend()

plt.show()
```

# Complete Example

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs")

plt.plot(x, x*1, label = "regular")
plt.plot(x, x*2, label = "squared")

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.legend()

plt.show()
```

# Save to File

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Save to File")

plt.plot(x, x*1, label = "regular")
plt.plot(x, x*2, label = "squared")

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.legend()

plt.savefig("complete_plot.png")

plt.show()
```

# Decoration with Plot Styles and Types

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs Colors")

plt.plot(x, 'y')
plt.plot(x + 1, 'g')
plt.plot(x + 2, 'r')

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Line Style")

plt.plot(x, '-')
plt.plot(x + 1, '*')
plt.plot(x + 2, '--')

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Marker Style")

plt.plot(x, 'D')
plt.plot(x + 1, 'o')
plt.plot(x + 2, 's')

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Customizations")

plt.plot(x, 'cD-')
plt.plot(x + 1, 'go')
plt.plot(x + 2, 'rs--')

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Customizations")

plt.plot(x, color="red", linestyle="dashdot", linewidth=3, marker="D", markersize=10)

plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```

```python
x = np.arange(1, 5)

plt.title("Matplotlib Graphs - Ticks")

plt.plot(x, color="red", linestyle="dashdot", linewidth=3, marker="D", markersize=10)
plt.xticks(range(len(x)), ['Pune', 'Mumbai', "Bangalore", "Delhi"])
plt.yticks(range(0, 8, 2))
plt.grid(False)
plt.xlabel('numbers')
plt.ylabel('derived')

plt.show()
```
